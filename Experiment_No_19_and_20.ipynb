```python
# Upload the Dataset
import pandas as pd
data=pd.read_csv('/content/covid_19_data.csv')
data
```

```python
#Removing Redundant Values
data=data.drop(['SNo','Last Update'],axis=1)
data.head()
```

```python
# Info of the given data
data.info()
```

```python
# To find the datatype of the given data
data['ObservationDate']=data['ObservationDate'].astype('datetime64[ns]')
data['Confirmed']=data['Confirmed'].astype('int64')
data['Deaths']=data['Deaths'].astype('int64')
data['Recovered']=data['Recovered'].astype('int64')




# Info of the given data
data.info()
```

```python
# Rechecking the data
data.head()
```

```python
# Active Case
data['Active']=data['Confirmed']-data['Deaths']-data['Recovered']
data.iloc[50:100]
```

```python
# last/current date
data['ObservationDate'].max()
```

```python
# Fetch the latest data
latest_data=data[data['ObservationDate']==data['ObservationDate'].max()]
latest_data.head()
```

```python
# Shape of the latest data
latest_data.shape
```

```python
# Country wise count
latest_data['Country/Region'].value_counts()
```

```python
# No of Countries
latest_data['Country/Region'].unique()
```

```python
# Count of Countries
latest_data['Country/Region'].nunique()
```

```python
# Count of the cases for each Country
countries=latest_data.groupby('Country/Region')[['Confirmed','Deaths','Recovered','Active']].sum()
countries=countries.reset_index()
countries
```

```python
#Count for a specific Country
countries[countries['Country/Region']=='India']
```

```python
countries[countries['Country/Region']=='Russia']
```

```python
# Plot on World Map
import plotly.express as px
```

```python
world_map=px.choropleth(countries,locations='Country/Region',locationmode='country names',color='Confirmed',color_continuous_scale='pubu',range_color=[0,10000000])
world_map
```

```python
# Top 20 countries having maximum confirmed and recovered
top=latest_data.groupby("Country/Region")[["Confirmed","Recovered"]].sum().reset_index()
top=top.sort_values("Confirmed",ascending=False)
top.head(20)
```
